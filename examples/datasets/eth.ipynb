# ETH Pedestrian Dataset

```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.ioff()
from IPython.display import HTML
```

```python
from mantrap.environment import Trajectron

from mantrap_evaluation.datasets import scenario_eth

env, _, trajectories_ground_truth = scenario_eth(Trajectron, ego_type=None, t_dataset=80.0, num_modes=4)
```

```python
agents = env.ados()
for ado in agents: 
    ado_id = ado.id
    plt.plot(ado.history[:, 0], ado.history[:, 1], "-", label=f"{ado_id}")
    
    ground_truth = trajectories_ground_truth[ado_id]
    plt.plot(ground_truth[:, 0], ground_truth[:, 1], "--")
    
    
plt.legend()
plt.show()
```
