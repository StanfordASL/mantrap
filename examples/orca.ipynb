# ORCA Solver

```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.ioff()
from IPython.display import HTML
```

```python
import mantrap
import mantrap_evaluation
import torch


env, goal, _ = mantrap_evaluation.scenarios.custom_surrounding(mantrap.environment.PotentialFieldEnvironment, 
                                                               ego_type=mantrap.agents.IntegratorDTAgent, dt=0.1)
solver = mantrap.solver.baselines.ORCASolver(env=env, goal=goal, is_logging=True)
solver.solve(time_steps=100)
HTML(solver.visualize_scenes())
```
