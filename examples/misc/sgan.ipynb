# SGAN Environment

```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.ioff()
```

```python
import mantrap
import torch


sgan = mantrap.environment.SGAN(ego_position=torch.zeros(2), ego_velocity=torch.rand(2))
sgan.add_ado(position=torch.tensor([4, 2]), velocity=torch.tensor([-1, -1]))
sgan.add_ado(position=torch.tensor([-4, 7]), velocity=torch.tensor([1, 0]))
sgan.add_ado(position=torch.tensor([-2, 6]), velocity=torch.tensor([-1, 0]))

samples = sgan.sample_wo_ego(t_horizon=7, num_samples=1)
sgan.visualize_prediction_wo_ego(t_horizon=5)
plt.show()
```
